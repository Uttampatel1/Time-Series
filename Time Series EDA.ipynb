### Time Resampling

```python
df_tesla.resample(rule='A').max()
```

```python
df_tesla.resample(rule='A').min()
```

```python
## year and frequency
df_tesla.resample(rule='A').max()['Open'].plot()
```

```python
## quarter and frequency

df_tesla.resample(rule='QS').max()['High'].plot()
```

```python
## Bissness day and frequncy

df_tesla.resample(rule='B').max()['High'].plot()
```

```python
df_tesla.resample(rule='BQS').max()['High'].plot()
```

```python
df_tesla.resample(rule='BA').max()['High'].plot()
```

```python
## ploatting
```

```python

df_tesla['Open'].resample(rule='BA').mean().plot(kind='bar',figsize=(14, 5))
```

```python

df_tesla['Open'].resample(rule='M').max().plot(kind='bar',xlim=['2018-01-01','2023-01-01'],figsize=(14, 5))
```

```python
df_tesla['High'].rolling(window=10).mean().head(20)
```

```python
df_tesla['Open : 30 day of moving average'] = df_tesla['Open'].rolling(window=30).mean()
```

```python
df_tesla[['Open','Open : 30 day of moving average']].plot(figsize=(14, 5),xlim=['2020-01-01','2023-01-01'])
```

```python
df_tesla['Open: 44 day of moving average'] = df_tesla['Open'].rolling(window=44).mean()
```

```python
df_tesla[['Open','Open : 30 day of moving average','Open: 44 day of moving average']].plot(figsize=(14, 5),xlim=['2020-01-01','2023-01-01'])
```
